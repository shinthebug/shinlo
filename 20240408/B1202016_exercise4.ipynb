```python
h=float(input('請輸入身高公分'))
w=float(input('請輸入體重公斤'))
h*=0.01
p=1/h
bmi=w*p*p
print(bmi)
if bmi>=27:
 print("肥胖")
elif bmi>=24:
 print("過重")
elif bmi>=18.5:
 print("適中")
else:
 print('過輕')
```

```python
a=float(input('請輸入第一個數字'))
c=input('請輸入運算符號')
b=float(input('請輸入第二個數字'))
if c=="+":
 print(a+b)
elif c=="-":
 print(a-b)
elif c=="*":
 print(a*b)
else:
 print(a/b)
```

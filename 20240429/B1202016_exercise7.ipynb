```python
a='*'
def f(h):
  l=[]
  for i in range(0,h):
    l.append(" "* (h-i-1)+a*(2*i+1)+" "*(h-i-1))
  for j in range(5):
      l.append(" "*(h-2)+a*3+" "*(h-2))
  return l
h=input('聖誕樹高度：')
l=(f(int(h)))
for x in l:
  print(x)
```

```python
height = input("請輸入身高")
weight = input("請輸入體重")

def BMI(height,weight):
  BMI = weight/height**2
  if BMI<18.5:
    ans = "too thin"
  elif BMI<24 and BMI >=18.5:
    ans = "moderate"
  elif BMI>24 and BMI<27:
    ans="too fat"
  else:
    ans="obesity"
  return BMI, ans

try:
  height = float(height)
  weight = float(weight)
  height /= 100
  BMI, ans = BMI(height,weight)
  print(f'BMI:{BMI}')
  print(f"體位:{ans}")
except:
  print("輸入內容錯誤")
```
